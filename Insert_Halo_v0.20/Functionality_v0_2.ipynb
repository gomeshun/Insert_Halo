```python
from IH_main import *
from IH_data_analysis import Insert_Halo_Data_Analysis
```

```python
# From version 0.20 of Insert Halo and onward one can no longer call the main function directly. Instead it is now required to 
# make an Insert_Halo class first:
IH = Insert_Halo()

# And then call the main function located in this class:
IH.run_insert_halo(zlens = 0.5, zsource = 2.5, DM_type = "CDM", Mhost = 13.3, geometry_shape = "DOUBLE_CONE", cone_angle_arcsec = 10,
                      idle_run = False, randomization = False, seed = None, LOS_normalization = 1, 
                        host_halo_scale_factor = 0.88, redshift_scale_factor = 1.7, fix_mass_range_sh = True, msub_min = 6, 
                        correlated_clustering = True, nph_subhalos = None)
```

```python
# The reason for this change is to allow Insert Halo to be rerun easily, without having to rerun Sashimi. To this end the 
# rerun_insert_halo function was added to this class as well, which is effectively the same as run_insert_halo, but allows
# the program to run without having to run Sashimi again and instead it simply uses the already existing Sashimi data from
# the previous run.

IH.rerun_insert_halo(zlens = 0.5, zsource = 2.5, DM_type = "CDM", Mhost = 13.3, geometry_shape = "DOUBLE_CONE", cone_angle_arcsec = 10,
                      idle_run = False, randomization = False, seed = None, LOS_normalization = 1, 
                        host_halo_scale_factor = 0.88, redshift_scale_factor = 1.7, fix_mass_range_sh = True, msub_min = 6, 
                        correlated_clustering = True, nph_subhalos = None)
```

```python
# Of course we could change some settings in the rerun_insert_halo function, which would mean the Sashimi data is no longer
# correct. If this is the case rerun_insert_halo does run Sashimi again, with altered parameters. Here is an example where
# we changed the redshift of the lens:

IH.rerun_insert_halo(zlens = 1.0, zsource = 2.5, DM_type = "CDM", Mhost = 13.3, geometry_shape = "DOUBLE_CONE", cone_angle_arcsec = 10,
                      idle_run = False, randomization = False, seed = None, LOS_normalization = 1, 
                        host_halo_scale_factor = 0.88, redshift_scale_factor = 1.7, fix_mass_range_sh = True, msub_min = 6, 
                        correlated_clustering = True, nph_subhalos = None)
```
