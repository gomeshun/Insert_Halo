```python

print('realization contains '+str(len(new_cdm.halos))+' halos.')
cdm_halo_masses = [halo.mass for halo in new_cdm.halos]
N, m = np.histogram(cdm_halo_masses, bins=np.logspace(5, 10, 10))

print(type(cdm_halo_masses[0]))

fig = plt.figure(1)
ax = plt.subplot(111)
ax.loglog(m[0:-1], N)
ax.set_xlabel('Halo mass '+r'$M_{\odot}$', fontsize=16)
ax.set_ylabel('n(M)', fontsize=16)

fig = plt.figure(2)
fig.set_size_inches(14,12)
ax = plt.subplot(111, projection='3d')
new_cdm.plot(ax)


save_fig = True
if save_fig:
    plt.savefig(f"/home/stefan/Documents/uva/Thesis/Images/Insert_halo/subhalo_spatial_distribution_3d.pdf", bbox_inches = "tight")
```
